```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import datetime
import pandas as pd
```

```python

extension='hea'
path='physionet.org/files/fantasia/1.0.0'
files=glob.glob(path+"/*."+extension)
ages=[]
sexs=[]
for file in files:
    with open(file, 'r') as f:
        file_content=f.readlines()
        anagraphic=file_content[-1].strip().split(' ')
        if anagraphic[1]=='?':
            age,sex=0, anagraphic[4]
        else:
            age, sex=int(anagraphic[2]), anagraphic[4]
        ages.append(age)
        sexs.append(sex)
ages=np.array(ages)
sexs=np.array(sexs)

age_bin=np.arange(np.min(ages)-5, np.max(ages)+5, 2)

plt.hist([ages[(sexs=='M')],ages[(sexs=='F')]],
        age_bin, stacked=True, label=['M','F'], color=['b','r'], align='left')
#plt.hist(ages[(ages<50) & (sexs=='F')], age_bin, stacked=True, label='Young F', color='indianred', align='left')
#
#plt.hist(ages[(ages>50) & (sexs=='M')], age_bin, stacked=True, label='Elderly M', color='darkblue', align='left')
#plt.hist(ages[(ages>50) & (sexs=='F')], age_bin, stacked=True, label='Elderly F', color='firebrick', align='left')
plt.axvline(50, linestyle='dashed', color='k')
plt.text(49, 3, 'Young', va='center', ha='right')
plt.text(51, 3, 'Elderly', va='center', ha='left')


plt.legend(title='Sex')
plt.grid(linestyle='dotted')
plt.xlabel('Ages [years]')
date=datetime.datetime.now().strftime("%d %b %H:%M")
ax=plt.gca()
ax.annotate('Generated on '+date, size=6, xy=(0,1+0.01), xycoords='axes fraction', color='grey', ha='left')
plt.yticks(np.arange(0,7), np.arange(0,7))
plt.ylabel('Counts')
plt.tight_layout()
#ax.annotate('(1.) Fantasia dataset', xy=(0.5, -0.15), xycoords='axes fraction', ha='center')
plt.savefig('images/fantasia_patient_distribution.png')
```
